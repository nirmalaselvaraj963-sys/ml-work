```python
#importing the Libraies
import numpy as np
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
X=dataset[['Age', 'EstimatedSalary','Gender_Male']]
y=dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define parameter grid for RandomForestClassifier
param_grid = {
    'n_estimators': [100, 200, 300],        # number of trees
    'criterion': ['gini', 'entropy', 'log_loss'],  # splitting criteria
    # 'max_depth': [None, 10, 20, 30],        # maximum depth of trees
    # 'min_samples_split': [2, 5, 10],        # min samples required to split
    # 'min_samples_leaf': [1, 2, 4],          # min samples at leaf
    # 'max_features': ['sqrt', 'log2', None]  # number of features to consider at split
}

# Initialize GridSearchCV with RandomForestClassifier
grid = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    refit=True,
    verbose=3,
    n_jobs=-1,
    scoring='f1_weighted'   # suitable for imbalanced classes
)

# Fit on training data
grid.fit(X, y)
```

```python
from sklearn.model_selection import cross_val_predict

# Use the best estimator from GridSearchCV
best_model = grid.best_estimator_

# Perform cross-validated predictions
y_pred = cross_val_predict(best_model, X, y, cv=5, n_jobs=-1)
```

```python
import pickle

filename = "RandomForest_CV_best_model.sav"
pickle.dump(grid.best_estimator_, open(filename, 'wb'))
loaded_model = pickle.load(open(filename, 'rb'))
y_pred = loaded_model.predict(X)
```
